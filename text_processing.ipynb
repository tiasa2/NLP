```
from nltk.corpus import movie_reviews
```

```
print(movie_reviews.categories())
```

```
print(len((movie_reviews.fileids('pos'))))
print(' ')
print(movie_reviews.fileids('pos'))
```

```
print(len((movie_reviews.fileids('neg'))))
print(' ')
neg_rev=movie_reviews.fileids('neg')
```

```
rev=nltk.corpus.movie_reviews.words('pos/cv010_29198.txt')
print(rev)
```

```
rev_list=[]
for rev in neg_rev:
  rev_text_neg=rev=movie_reviews.words(rev)
  review_one_string=' '.join(rev_text_neg)
  review_one_string=review_one_string.replace(' ,',',')
  review_one_string=review_one_string.replace('\' ',"'")
  review_one_string=review_one_string.replace('\'',"'")
  rev_list.append(review_one_string)
```

```
pos_rev=movie_reviews.fileids('pos')
for rev in pos_rev:
  rev_text_pos=rev=movie_reviews.words(rev)
  review_one_string=' '.join(rev_text_neg)
  review_one_string=review_one_string.replace(' ,',',')
  review_one_string=review_one_string.replace('\' ',"'")
  review_one_string=review_one_string.replace('\'',"'")
  rev_list.append(review_one_string)
```

```
len(rev_list)
```

```
neg_targets=np.zeros((1000,),dtype=np.int)
pos_targets=np.ones((1000,),dtype=np.int)
```

```
target_list=[]
for n_tar in neg_targets:
  target_list.append(n_tar)
for p_tar in pos_targets:
  target_list.append(p_tar)
```

```
len(target_list)
```

```
y=pd.Series(target_list)
type(y)
```

```
count_vec=CountVectorizer(lowercase=True,stop_words='english',min_df=2)
```

```
X_count_vec=count_vec.fit_transform(rev_list)
```

```
X_count_vec.shape
```

```
X_names=count_vec.get_feature_names()
X_names
```

```
X_count_vec=pd.DataFrame(X_count_vec.toarray(),columns=X_names)
```

```
X_count_vec.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```
X_train,X_test,y_train,y_test=train_test_split(X_count_vec,y,test_size=0.25,random_state=5)
```

```
X_train.shape,X_test.shape
```

```
from sklearn.naive_bayes import GaussianNB
```

```
gnb=GaussianNB()
y_pred_gnb=gnb.fit(X_train,y_train).predict(X_test)
```

```
from sklearn.naive_bayes import MultinomialNB
```

```
cv=MultinomialNB()
y_pred_cv=cv.fit(X_train,y_train).predict(X_test)
```

```
print(metrics.accuracy_score(y_test,y_pred_cv))
```

```
score=confusion_matrix(y_test,y_pred_cv)
score
```
